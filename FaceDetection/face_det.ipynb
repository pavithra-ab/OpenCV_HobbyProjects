```python
import cv2
```

```python
img = cv2.imread(r'F:\DL Projects\OpenCVProjects\FaceDetection\srk.jpg')
img
```

```python
cv2.imshow("Image",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
con_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
face_cas = cv2.CascadeClassifier(r'F:\DL Projects\OpenCVProjects\FaceDetection\haarcascade_frontalface_default.xml')
face_cas
```

```python
#it is good to process as gray scale
img_gray = cv2.cvtColor(con_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
cv2.imshow('GRAY IMG',img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
face = face_cas.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=6)
face
```

```python
len(face)
```

```python
plt.imshow(con_img)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
video = cv2.VideoCapture(r'F:\DL Projects\OpenCVProjects\FaceDetection\video.mp4')
while True: #how long the video exists this loop should run
    success,frame = video.read() #success and frame are 2 variables.when frame var has a value, success becomes True
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) #to convert to grayscale
    faces = face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    print(faces)
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'): # conditions, When if i press q it will wait for 1ms and then break
        break
video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(r'F:\DL Projects\OpenCVProjects\FaceDetection\video.mp4')
while True: #how long the video exists this loop should run
    success,frame = video.read() #success and frame are 2 variables.when frame var has a value, success becomes True
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) #to convert to grayscale
    faces = face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    #print(faces)

    #to draw the rectangle
    for (x,y,w,h) in faces:
           cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,1),3)
           cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'): # conditions, When if i press q it will wait for 1ms and then break
           break
video.release()
cv2.destroyAllWindows()
```

```python
#using webcam
video = cv2.VideoCapture(0)
while True:
    success, frame = video.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    #print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,1),3)
        cv2.imshow('webcam',frame)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
