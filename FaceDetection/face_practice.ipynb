```python
import cv2
```

```python
img = cv2.imread(r"F:\DL Projects\OpenCVProjects\FaceDetection\bts.png")
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
con_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
face_cas = cv2.CascadeClassifier(r"F:\DL Projects\OpenCVProjects\FaceDetection\haarcascade_frontalface_default.xml")
face_cas
```

```python
img_gray = cv2.cvtColor(con_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
cv2.imshow("GRAY Image",img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
face = face_cas.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=2)
face
```

```python
plt.imshow(con_img)
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),4)
```
