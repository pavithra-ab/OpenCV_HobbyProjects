```python
import cv2
```

```python
img = cv2.imread(r'F:\DL Projects\OpenCVProjects\FaceDetection\image.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
face_cas = cv2.CascadeClassifier(r"F:\DL Projects\OpenCVProjects\FaceDetection\haarcascade_frontalface_default.xml")
face_cas
```

```python
face = face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=6)
face
```

```python
len(face)
```

```python
con_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
for x,y,w,h in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(con_img)
```

```python
video = cv2.VideoCapture(r"F:\DL Projects\OpenCVProjects\FaceDetection\video.mp4")
while True:
    success,frame = video.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face = face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    for x,y,w,h in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.imshow("Video",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
