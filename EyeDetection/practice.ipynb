```python
import cv2
```

```python
img = cv2.imread(r'F:\DL Projects\OpenCVProjects\EyeDetection\namjin.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
eye_cas = cv2.CascadeClassifier(r'F:\DL Projects\OpenCVProjects\EyeDetection\haarcascade_eye.xml')
eye_cas
```

```python
eye = eye_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
eye
```

```python
con_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
for x,y,w,h in eye:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),3)
plt.imshow(con_img)
```

```python
video = cv2.VideoCapture(r'F:\DL Projects\OpenCVProjects\EyeDetection\video.mp4')
while True:
    success,frame = video.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    eye = eye_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=14)
    for x,y,w,h in eye:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.imshow("IMG",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
