```python
import cv2
```

```python
img = cv2.imread(r'F:\DL Projects\OpenCVProjects\EyeDetection\srk.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
eye_cas = cv2.CascadeClassifier(r'F:\DL Projects\OpenCVProjects\EyeDetection\haarcascade_eye.xml')
eye_cas
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
eye = eye_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=6)
eye
```

```python
con_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
for (x,y,w,h) in eye:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(con_img)
```

```python
img1 = cv2.imread(r'F:\DL Projects\OpenCVProjects\EyeDetection\eyeimg.jpg')
plt.imshow(img1)
```

```python
plt.imshow(con_img1)
```

```python
#video
video = cv2.VideoCapture(r'F:\DL Projects\OpenCVProjects\EyeDetection\video.mp4')
while True:
    success,frame = video.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    eye = eye_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=16)

    for (x,y,w,h) in eye:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
        cv2.imshow('Video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
